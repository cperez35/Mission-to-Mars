```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager

# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# Visit the Mars Facts site
url = 'https://redplanetscience.com/'
browser.visit(url)

html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
#Scrape titles
titles= html_soup.find_all('div',class_='content_title')
for title in titles:
    print(titles)
```

```python
#scrape previews
previews= html_soup.find_all('div',class_='article_teaser_body')
for preview in previews:
    print(previews)
```

```python
text_elems = html_soup.find_all('div', class_='list_text')
news_items = []
for pane in text_elems:
    title = pane.find('div', class_= 'content_title').text
    preview = pane.find('div', class_= 'article_teaser_body').text
    news_item = {}
    news_item ['title'] = title
    news_item ['preview'] = preview
    news_items.append(news_item)
news_items
```

```python
import json
```

```python
mars_dict = json.dumps(news_items)
print(type(mars_dict))
print(mars_dict)
```
